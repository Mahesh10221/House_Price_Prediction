```python
def label_encoding(original, replace, l1):
    original = list(original)
    replace = list(replace)
    l1 = list(l1)
    new_list = l1.copy()

    for i in range(len(new_list)):
        if new_list[i] in original:
            index = original.index(new_list[i])
            new_list[i] = replace[index]

    return pd.Series(new_list)
```

```python

def append_row_to_df(df, new_row_values):
    """
    Appends a new row to a dataframe with values given in new_row_values list.

    Parameters:
    df (pandas.DataFrame): The dataframe to which the new row is to be appended.
    new_row_values (list): A list of values to be appended as a new row. The length of the list should be equal to the number of columns in the dataframe.

    Returns:
    pandas.DataFrame: The dataframe with the new row appended.
    """

    # Check if the length of new_row_values is equal to the number of columns in the dataframe
    if len(new_row_values) != len(df.columns):
        print("Error: The length of new_row_values does not match the number of columns in the dataframe.")
        return df

    # Create a dictionary with keys as column names and values as the new row values
    new_row_dict = dict(zip(df.columns, new_row_values))

    # Append the new row to the dataframe and reset the index
    df = df.append(new_row_dict, ignore_index=True)

    return df
```

```python
def replace_values(lst, old_vals, new_vals):
    """
    Replaces all occurrences of the old values in the given list with the corresponding new values.
    """
    for i in range(len(lst)):
        for j in range(len(old_vals)):
            if lst[i] == old_vals[j]:
                lst[i] = new_vals[j]
    return lst
```
